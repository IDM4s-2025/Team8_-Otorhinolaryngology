## Medical Specialty: Otorhinolaryngology

For this project, a medical specialty, Otolaryngology, has been assigned; using this medical specialty an expert medical diagnostic system was designed. Below is information on 12 diseases related to the specialty along with symptoms supported by medical references.

### Diseases: 

### Acoustic neuroma

An acoustic neuroma is a noncancerous tumor on the main nerve that runs from the inner ear to the brain. This nerve is called the vestibular nerve. Branches of the nerve directly affect balance and hearing. Pressure from an acoustic neuroma can cause hearing loss, ringing in the ears, and balance problems.

The known syntoms are:
- Hearing loss
- Ringing in the affected ear, known as tinnitus.
- Loss of balance or feeling unsteady.
- Dizziness.
- Facial numbness and, very rarely, weakness or loss of muscle movement.

(Mayo Clinic, 2023)

### Otitis externa

Acute otitis externa (AEO), also known as swimmer's ear or swimming pool ear, is a diffuse inflammation of the external auditory canal (EAC) that can extend to the auricle or tympanic membrane. Acute bacterial infection is the most common cause of otitis externa.
- Earache
- Itching on the ear
- Blocked ear

(Cano et al., 2024)

### Sleep apnea: 

Sleep apnea is a potentially serious sleep disorder in which breathing repeatedly stops and starts. If you snore loudly and feel tired even after a full night's sleep, you may have sleep apnea.
The main symptoms are:
- Loud snoring.
- Episodes where your breathing stops while you sleep (something that should be reported by someone else).
- Gasping for breath in your sleep
- Waking up with a dry mouth.
- Headache in the morning
- Insomnia
- Excessive daytime sleepiness, known as hypersomnia
- Difficulty paying attention while awake
- Irritability

(Mayo Clinic, 2023)

### Otitis Media

Ear Infection - Acute Otitis Media, “is an infection of the air-filled space behind the eardrum (Mayo Clinic, s.f.).”   Children are the most likely to get this kind of infection, although sometimes adults also get it. The common symptoms for children are:

- Ear pain
- Tugging at an ear
- Trouble sleeping
- Crying more than usual
- Fussiness
- Trouble hearing or replying to sounds
- Loss of balance
- Fever
- Fluid coming from the ear
- Headache
- Loss of appetite

The symptoms for adults are:
- Ear Pain
- Fluid coming from the ear
- Trouble hearing

(Mayo Clinic, 2025)


### Meniere's Disease 

Meniere's disease is a disorder of the inner ear that can lead to episodes of severe dizziness, known as vertigo, and hearing loss. In most cases, this condition only impacts one ear.

- Frequent dizziness: a sensation of spinning that begins and ends suddenly, which can last between 20 minutes and 12 hours, and sometimes more than 24 hours. Severe dizziness can cause nausea.
- Hearing loss: which may be intermittent at first and, over time, more permanent.
- Tinnitus: Ringing, whistling, roaring, hissing, or ringing in the ear.
- Feeling of fullness in the ear: A feeling of pressure or congestion in the affected ear.

(Mayo Clinic, 2024)

### Epistaxis

Epistaxis or nosebleed is any bleeding originating from the nasal passages.  This bleeding may be caused by trauma, irritation or dryness of the lining of the nose. The main sign or symptom of epistaxis is nosebleeds. At first, epistaxis is usually scanty and of limited duration. In more severe cases, blood may flow out of one nostril and/or become bilateral as blood flows through the nostrils.
Other symptoms of epistaxis are:
- Sweating.
- Dizziness.
- Nausea and vomiting, due to swallowing blood.
- Tachycardia.
- Hypotension.

(Top Doctors, 2016)

### Sinusitis

It's the inflammation of the sinuses (spaces in the forehead, cheeks and nose). It's usually caused by infection and it clears up in less than 4 weeks. Sinusitis blocks the sinuses with fluid, causing the following symptoms:

- Pain, swelling and tenderness around cheeks, eyes or forehead.
- Blocked or runny nose.
- Reduce the sense of smell.
- Green or yellow mucus.
- High temperature.
- Headache
- Toothache
- Pressure in the ears
- Cough
- Bad breath

(NHS, 2024)

### Allergic rhinitis

Causes cold-like symptoms, but allergic rhinitis is not caused by a virus, but by an allergic response to a harmless substance from inside or outside that the body identifies as harmful (allergen).
Symptoms of allergic rhinitis may include the following:

- Runny nose and nasal congestion.
- Red, watery, itchy eyes.
- Sneezing
- Coughing
- Itching of the nose, roof of the mouth or throat.
- Mucus running down the back of the throat, called postnasal drip.
- Bluish, swollen skin under the eyes, known as allergic black eyes.
- Extreme tiredness and fatigue, often due to lack of sleep.

(Mayo Clinic, 2024)

### Nasal Polyps

Nasal polyps are painless, noncancerous growths that develop inside the nose or the hollow areas within the facial bones (sinuses). They are associated with inflammation of the inside of the nose and sinuses, often linked to chronic sinusitis, which can last more than 12 weeks. These growths can range from small and asymptomatic to large and obstructive. The common symptoms are: 

- Congestion and runny nose
- Mucus running down the throat (postnasal spike)
- Loss of smell and taste
- Facial pain or headache
- Tooth pain
- Feeling of pressure on the forehead and face
- Snoring
- Has benign polyps 

(Mayo Clinic, 2023)

### Tonsillitis

It's the inflammation of the tonsils (back throat oval-shaped pads). It commonly occurs in children and during the mid teenage years. The symptoms that may appear are the following:

- Red, swollen tonsils.
- White or yellow coating or patches on the tonsils
- Sore throat
- Difficulty or painful to swallow
- Fever
- Enlarged, tender glands in the neck
- A scratchy, muffled or  throaty voice 
- Bad breath
- Stomachache
- Neck pain or stiff neck
- Headache
- Drooling
- Unusual fussiness

It's recommended to go to a doctor if the person has one of the following:
- Sore throat and fever
- Sore throat hasn't gone away in the last 24 to 48 hours.
- Painful or difficult to swallow
- Extreme weakness, fatigue or fussiness.

(Mayo Clinic, 2022)

### Pharyngitis

Pharyngitis, or sore throat, is discomfort, pain, or scratchiness in the throat that often causes pain when swallowing. Pharyngitis is caused by swelling of the back of the throat (pharynx), between the tonsils and the voice box (larynx). Most sore throats are caused by colds, flu, coxsackie virus, or mononucleosis.

Symptoms: 
- The main symptom is a sore throat.

Other symptoms may include:
- Fever
- Headache
- Muscle and joint aches
- Rash
- Swollen lymph nodes in the neck

(MedlinePlus, 2023)

### Laryngitis

Is the inflammation of the voice box (larynx) caused by overuse, irritation or infection. When this happens, the cords swell, distorting the sounds produced and resulting in a hoarse voice. In some cases, the voice becomes imperceptible. The laryngitis can be either short-term (acute) or long-term (chronic). 

In adults, the symptoms are:
- Hoarseness or loss of voice
- Weakness or muffled voice
- Tingling or rough feeling in the throat
- Sore throat
- Dry throat
- Dry cough

 (Mayo Clinic, 2022)


# Medical diagnosis expert system

This project shows a design of a otorhinolaryngology diagnosis expert system using *Experta* based on the diseases above. The decision tree below is the one we used for the algorithm:

<img src="DiagramES.jpg" alt="Decision Tree" width="600">

Based on the decision tree, a table was created to highlight the most frequent symptoms, as well as those that differentiate one disease from another:

| Dictionary | Fact Name              | Meaning                                         |
|--------|------------------------|-------------------------------------------------|
| `E`    | `ear_pain`             | Ear pain or discharge                           |
| `T`    | `tragal_tenderness`    | Tragus tenderness or canal swelling             |
| `M`    | `middle_fluid`         | Middle ear fluid with fever                     |
| `MN`   | `meniere_symptoms`     | Fullness, tinnitus, or vertigo                  |
| `NB`   | `nose_bleed`           | Active nosebleed                                |
| `FP`   | `facial_pain`          | Facial pain/pressure <12 weeks                  |
| `A`    | `allergy`              | Allergy symptoms (sneezing, clear discharge)    |
| `NBK`  | `nasal_blockage`       | Persistent nasal blockage                       |
| `TP`   | `throat_pain`          | Throat pain or discomfort when swallowing       |
| `EX`   | `exudate`              | White exudate on tonsils                        |
| `H`    | `hoarseness`           | Hoarseness or voice change                      |
| `AP`   | `apnea`                | Apnea, snoring, or daytime drowsiness           |

From the dictionary, the diagnoses are logical conclusions derived from symptom combinations, the next system (.py) assesses whether the patient exhibits specific symptom combinations. When characteristic symptoms such as ear pain, nasal congestion, or throat discomfort are detected, the corresponding rule is activated, and the system suggests the appropriate ENT diagnosis.

This is a guide for using the expert system to determine the diagnoses.
